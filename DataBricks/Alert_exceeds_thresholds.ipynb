```python
import requests
 
logic_app_url = "https://prod-08.uaenorth.logic.azure.com:443/workflows/837ba73752c342f79b098f3da46e44fd/triggers/When_an_HTTP_request_is_received/paths/invoke?api-version=2016-10-01&sp=%2Ftriggers%2FWhen_an_HTTP_request_is_received%2Frun&sv=1.0&sig=TtBI-7viu2LFzsBD4O5nW8envcLdfZFkTuZduXybKys"
 
num_vehicles = spark.sql("""
SELECT COUNT(vehicle_id) AS num_vehicles
FROM traffic.traffic_insights.Analyze_vehicle_silver
WHERE alert = 'Overspeed'
""").collect()[0]["num_vehicles"]
 
if num_vehicles > 20:
    response = requests.post(
        logic_app_url,
        json={"num_vehicles": num_vehicles}
    )
    print("Alert sent!", response.status_code)
else:
    print("No alert needed. Vehicles:", num_vehicles)
```
